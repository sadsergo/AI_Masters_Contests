### Решение
#### Этап 1

Так как это проблема стохастического анализа, то есть смысл её попробовать решить через метод Монте-Карло. 

<br> Условно можно выделить фигуры: треугольник, квадрат, пятиугольник и тд и промоделлировать эксперимент: много раз запускать случайные маршруты по сетке и смотреть, куда человек в итоге за 96 шагов пришел, посчитать частоту прихода и выбрать наиболее вероятный вариант.

Вероятней всего, такой алгоритм будет работать очень долго, особенно, если можно ещё ходить назад, то....может, через месяц посчитает. 

#### Этап 2

Рассмотрим 2 случая:
1. Если можно ходить назад: тогда, так как можно в любом узле пойти назад, то, например, из входа идем в какой-то узел, оттуда назад ко входу, так 4 раза, после четвертого, опять идем в какой-то узел, из этого узла в другой, между этими узлами перемещаемся нужное количество раз и потом обратно возвращаемся в пятый раз ко входу. Так как турист случайно выбирал звенья, то такой маршрут может быть осуществлен. 
<br> Тогда можно вообще выбрать ячейку любой формы!!!
2. Если нельзя ходить назад: тогда поступим так, перемещение из входа в узел - 1 ход, и назад из какого-то узла ко входу - тоже 1 ход. Перемещаемся из входа в другую ячейку и из нее тоже в какую-то. Тогда остается 92 хода (2 хода вперед и 2 хода назад, чтобы вернуться обратно ко входу). Аналогично с первым случаем, ходим по ячейке, содержащей вход, 4 раза, это 4 * k ходов, k - число вершин. Тогда общая формула подсчета всех ходов: $ 1 + 1 + 1 + 1 + 4 * k + n * k = 96$, n - число обходов какого-то другого k-гранника, у которого есть вершина, соединенная со входом. 
<br><br> Получается $k * (4 + n) = 92$. Делители 92 = {2, 4, 23, 46}. Тогда нам подходит k = 4, n = 19. 

Значит, если ячейка - квадрат, то турист мог 5 раз прийти к выходу.
