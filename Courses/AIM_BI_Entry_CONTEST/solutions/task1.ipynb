# Космонавты (Data mining/economic modeling)
### Описание решения
1. Проанализировать, сколько нужно новых рабочих и на какие должности, исходя из стадии проекта
2. Рассмотреть несколько российских университетов, где можно обучиться выбранным профессиям, и сравнить их по стоимости обучения
3. Выводы

### Решение

Для оценки потребности в специалистах каждого профиля надо учитывать следующие показатели: 
1. Планы компании по росту бизнеса (на текущий момент 2 успешных спутника при 1000 человек в штате, из которых 750 инженеры). То есть, необходимо оценить за какой временной период планируется увеличение количества спутников и насколько, и исходя из этого определять потребность в дополнительных кадрах.
2. Количество текущих сотрудников компании, которые потенциально покинут компанию в ближайшем будущем (процент ротации по выбытию персонала). 

#### Этап 1

Так как надо найти специалистов для разработки и внедрения современных космических технологий, то есть смысл рассмотреть компании, занимающиеся тем же самым и определить в процентном соотношении, сколько надо нанять новых сотрудников.

Попробуем на основе данных другой компании найти оптимальное количество новых работников для нас:
#### $\frac{\text{known open positions}} {\text{known employers count}} = \frac{\text{unknown open positions}} {1000}$

#### SpaceX 
В компании работает примерно 13000 [сотрудников](https://www.expressnews.com/business/article/spacex-s-starlink-poised-record-revenue-2024-19450589.php), на сайте открытых позиций примерно 735 [объявлений](https://www.spacex.com/careers/jobs/index.html?). Таким образом, отношение количества открытых вакансий к общему количеству работников:  $\frac{735} {13000} \approx 0.06$. Тогда приблизительное количество новых сотрудников компании Х равно 60

#### Расчет стоимости

```python
cost = [
    [262, 127, 430, 148],
    [329, 277, 127, 176],
    [270, 200, 277, 230],
    [100, 176, 216, 127],
    [278, 329, 150, 208],
    [260, 179, 168, 291],
    [300, 207, 288, 172],
    [98, 310, 329, 250],
    [442, 260, 519, 340],
    [381, 154, 350],
]

names = np.array(
[
    "ИТС",
    "КТЭС",
    "РСК",
    "П",
    "ЛТЛТ",
    "Р",
    "СУДН",
    "ИВТ",
    "ПМиФ",
    "ВПЭУ",
]
)

employers_count = np.array([11, 60, 3, 3, 1, 3, 3, 8, 6, 2])
```

##### 1. Средняя
Можно рассчитать среднюю цену за обучение и, исходя из количества сотрудников для найма, вычислить общую стоимость

```python
mean_costs_list = [np.mean(q) for q in cost]
mean_cost = np.sum(np.array(mean_costs_list) * employers_count)

plt.bar(names, mean_costs_list)
plt.title('Средняя стоимость обучения на каждой специальности')
plt.legend(['Стоимость (тыс.)'])
plt.show()
```

Прикладная математика и физика оказалась самой дорогой, в то время, как Приборостроение самое дешевое, теперь можно посмотреть, сколько будет стоить обучение для конкретного количества человек, нужных компании Х

```python
mean_costs_empl = np.array(mean_costs_list) * employers_count

plt.bar(names, mean_costs_empl)
plt.title("Средняя стоимость обучения на каждой специальности с учетом количества человек")
plt.legend(["Стоимость (тыс.)"])
plt.show()
```

```python
print(f"Средняя стоимость обучения 100 людей: {mean_cost}")
```

##### Таким образом, средняя стоимость обучения 100 людей - 24036.75 тысяч рублей в год

##### 2. Минимальная

```python
min_costs_list = [np.min(q) for q in cost]
min_cost = np.sum(np.array(min_costs_list) * employers_count)

plt.bar(names, min_costs_list)
plt.title("Минимальная стоимость обучения на каждой специальности")
plt.legend(["Стоимость (тыс.)"])
plt.show()
```

```python
min_costs_empl = np.array(min_costs_list) * employers_count

plt.bar(names, min_costs_empl)
plt.title(
    "Минимальная стоимость обучения на каждой специальности с учетом количества человек"
)
plt.legend(["Стоимость (тыс.)"])
plt.show()
```

```python
print(f"Минимальная стоимость обучения 100 людей: {min_cost}")
```

##### Вывод: минимальная стоимость обучения 100 людей - 13739 тысяч рублей в год

##### 3. Максимальная

```python
max_costs_list = [np.max(q) for q in cost]
max_cost = np.sum(np.array(max_costs_list) * employers_count)

plt.bar(names, max_costs_list)
plt.title("Максимальная стоимость обучения на каждой специальности")
plt.legend(["Стоимость (тыс.)"])
plt.show()
```

```python
max_costs_empl = np.array(max_costs_list) * employers_count

plt.bar(names, max_costs_empl)
plt.title(
    "Максимальная стоимость обучения на каждой специальности с учетом количества человек"
)
plt.legend(["Стоимость (тыс.)"])
plt.show()
```

```python
print(f"Максимальная стоимость обучения 100 людей: {max_cost}")
```

##### Вывод: максимальная стоимость обучения 100 людей - 34559 тысяч рублей в год
